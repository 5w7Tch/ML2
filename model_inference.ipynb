```python


df1 = pd.read_csv('/kaggle/input/testdata/test_transaction.csv')
df2 = pd.read_csv('/kaggle/input/testdata/test_identity.csv')
df = pd.merge(df1, df2, on='TransactionID', how='left')
df
```

```python
import mlflow
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load the MLflow model
logged_model = 'runs:/1ba8aaaa15b54a90953cbe572a645c69/model'
loaded_model = mlflow.pyfunc.load_model(logged_model)

# Retrieve the list of features used in training
run_id = '1ba8aaaa15b54a90953cbe572a645c69'  # Replace with your actual run ID if necessary
logged_features = mlflow.get_run(run_id).data.params['features']

# Clean the feature list by removing any leading/trailing spaces and quotes
logged_features = logged_features.replace("'", "").replace("[", "").replace("]", "").split(", ")

```

```python

# Check if all logged features are in the dataframe
print("Columns in df:", df.columns)

print("Logged features from MLflow:", logged_features)

missing_features = [feature for feature in logged_features if feature not in df.columns]
print("Missing features:", missing_features)

for col in missing_features:
    df[col] = 0  

X_test = df[logged_features] 

non_numeric_columns = X_test.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_columns)
```

```python

label_encoders = {}
for col in non_numeric_columns:
    le = LabelEncoder()
    X_test[col] = le.fit_transform(X_test[col])
    label_encoders[col] = le  

scaler = StandardScaler()
X_test_scaled = scaler.fit_transform(X_test)  

predictions = loaded_model.predict(X_test_scaled)

df['isFraud'] = predictions

print(df[['TransactionID', 'isFraud']].head())
```

```python
predictions_df = df[['TransactionID','isFraud']]  # Extract TransactionID and predictions
predictions_df.to_csv('new_new_predictions.csv', index=False) 
```
